# Ancestral sequence reconstruction in Python

Imports for tutorial.

```python
import phylopandas as pd
import dendropy as d
import pyasr
import toytree
```

Reconstruct a tree in a single line of Python code.

```python
# Use phylopandas to read a set of ancestor.s
df_seqs = pd.read_fasta('test.fasta')

# Use dendropy to read in tree.
tree = d.Tree.get(path='tree.newick', schema='newick')

# Reconstruct nodes in tree.
tree, df_seqs, df_anc = pyasr.reconstruct(df_seqs, tree, working_dir='test', alpha=1.235)
```

Write out ancester dataframe.

```python
df_anc.to_csv('ancestors.csv')
```

You can visualize your tree using the `toytree` package.

```python
# Get a toytree object.
tree_to_draw = toytree.tree(tree.as_string(schema='newick'))
node_labels = tree_to_draw.get_node_values('name', show_root=False, show_tips=False)

# Draw tree.
tree_to_draw.draw(width=400, height=400,
    tip_labels_align=True,
    use_edge_lengths=True,
    node_labels=node_labels,
                 node_size=18)
```
